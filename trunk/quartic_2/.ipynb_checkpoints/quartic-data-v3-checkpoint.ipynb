```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.serif'] = ['Times New Roman']
rcParams['font.size'] = 16
```

Import gap data (from DiagHam) and join with band geometry data (from BandGeometry.m)


```python
quartic_path = "~/DiagHam_Stability/trunk/quatic_2"

quartic_df = pd.read_csv(quartic_path+"June 2018/Many-body code/quartic-data-july/outfile.csv",
                        usecols=['UCarea','t2','gap'])
mixed_aniso_df = pd.read_csv(quartic_path+"June 2018/Many-body code/aniso-quartic-data/outfile.csv",
                        usecols=['UCarea','t2','gap'])
aniso_df = pd.read_csv(quartic_path+"June 2018/Many-body code/anisotropy-data/outfile-aniso.csv",
                        usecols=['UCarea','aniso','gap'])
rsquared_df = pd.read_csv(quartic_path+"June 2018/Many-body code/rsquared-data/outfile-rsquared.csv",
                        usecols=['UCarea','t2','gap'])
rsquared_aniso_df = pd.read_csv(quartic_path+"June 2018/Many-body code/rsquared-aniso/outfile-rsquared-aniso.csv",
                        usecols=['UCarea','aniso','gap'])
rsquared_mixed_df = pd.read_csv(quartic_path+"June 2018/Many-body code/rsquared-mixed-order/outfile-mixed.csv",
                        usecols=['UCarea','t2','gap'])
quartic_geom_df = pd.read_csv(quartic_path+"June 2018/Many-body code/quartic-geometry-data.csv")
mixed_geom_df = pd.read_csv(quartic_path+"June 2018/Many-body code/mixed-aniso-geometry-data.csv")
quartic_geom_df = quartic_geom_df.append(pd.read_csv(quartic_path+"June 2018/Many-body code/quartic-geometry-data-96.csv"))

quartic_full_df = pd.merge(quartic_df,quartic_geom_df,on=['UCarea','t2'])
#     #quartic_full_df=quartic_full_df.append(pd.merge(pd.read_csv("/Users/dbauer/Dropbox/quartic/June 2018/Many-body code/quartic-96-data/outfile96.csv"),
#        pd.read_csv("/Users/dbauer/Dropbox/quartic/June 2018/Many-body code/quartic-geometry-data-96.csv"),
#        on=['UCarea','t2']))
mixed_aniso_full_df = pd.merge(mixed_aniso_df,mixed_geom_df,on=['UCarea','t2'])
rsquared_full_df = pd.merge(rsquared_df,quartic_geom_df,on=['UCarea','t2'])
rsquared_mixed_full_df = pd.merge(rsquared_mixed_df,mixed_geom_df,on=['UCarea','t2'])
```
