Import gap data (from DiagHam) and join with band geometry data (from BandGeometry.m)


```python
# old
quartic_path = "~/DiagHam_Stability/trunk/quartic/"
quartic_boson_geom_df_old = pd.read_csv(quartic_path+"June 2018/Many-body code/quartic-boson-geometry.csv")
DiagHam_path = "~/DiagHam_Stability/trunk/FTI/src/Programs/FCI/"
quartic_boson_df_old = pd.read_csv(DiagHam_path+"quartic-bosons/outfile-boson-new.csv", 
                               usecols=['UCarea','t2','gap'])
boson_full_df_old = pd.merge(quartic_boson_df_old,quartic_boson_geom_df_old,on=['UCarea','t2'])

# new
quartic_2_path = "~/DiagHam_Stability/trunk/quartic_2/"
quartic_boson_df = pd.read_csv(quartic_2_path+"quartic-bosons-2/outfile-boson-new.csv", 
                               usecols=['UCarea','t2','gap'])
quartic_boson_geom_df = pd.read_csv(quartic_2_path+"quartic-boson-geometry.csv")
boson_full_df = pd.merge(quartic_boson_df,quartic_boson_geom_df,on=['UCarea','t2'])
```

```python
plt.figure(figsize=(10,10))
plt.scatter(boson_full_df.loc[boson_full_df['UCarea']==16]['traceIneq'],
            boson_full_df.loc[boson_full_df['UCarea']==16]['gap'],
            c=boson_full_df.loc[boson_full_df['UCarea']==16]['t2'],
           cmap=plt.cm.viridis, edgecolor='black', s=180)
#cbar=plt.colorbar()
#cbar.ax.set_title("t2")
plt.xlabel(r"$\langle T \rangle$",fontsize=26)
plt.xticks(fontsize=26)
plt.ylabel("Many-body gap",fontsize=26)
plt.yticks(fontsize=26)
plt.title(r"$\nu=1/2$ bosons, $\phi=1/16$, onsite interaction",fontsize=24)
plt.xlim(-0.0005,0.016)
plt.ylim(0.607,0.616)
plt.savefig("gap-v-trace-bosons-16-onsite.pdf",
            pad_inches=0,
            bbox_inches='tight')
```
